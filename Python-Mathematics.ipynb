### Mathematical Operations in Python


```python
# Addition 
c = 10 + 45
print(c)
```

```python
# Subtraction
d = 100 - 340    
print(d)
```

```python
# multiplication
e = 10 * 11
print(e)
```

```python
# division (floating point division)
f = 100 / 3
print(f, type(f))
```

```python
# division (floatiint point division)
f = 100 // 3
print(f, type(f))
```

```python
# exponentiation
g = 2 ** 3 # 2 to the power of 3
print(g)
```

```python
# Import math Library
import math

# Return the remainder of x/y
print (math.remainder(9, 2))
print (math.remainder(9, 3))
print (math.remainder(18, 4))
```

```python
print(math.sqrt(16)) # square root of 16
print(math.sqrt(25)) # square root of 25            
print(math.sqrt(36)) # square root of 2
```

```python
# Convert long lines into multiple lines by \
    
value_1 = 10
value_2 = 20
value_3 = 30   
value_4 = 40
value_5 = 50
value_6 = 60

# (Long line)
total = value_1 + value_2 + value_3 + value_4 + value_5 + value_6 
print(total)

# (Short lines)
 
total = value_1 + value_2 + \
        value_3 + value_4 + \
        value_5 + value_6


print(total)
```

### Order of Mathematical Operations 




```python
### Bad Code with Magic Numbers
laptop_price = 1000
final_price = (laptop_price - (laptop_price * 0.10)) * 1.13
print(final_price)

# ### Good Code without Magic Numbers
laptop_price = 1000
DISCOUNT_RATE = 0.10
TAX_RATE = 1.13
final_price = (laptop_price - (laptop_price * DISCOUNT_RATE)) * TAX_RATE
print(final_price)
 

 
```
