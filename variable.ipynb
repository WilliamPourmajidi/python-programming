### Float Variables 

```python
pi = 3.14 #float    
print(pi, type(pi))

i = 200
print(i, type(i))
# Casting 
float_i = float(i) #convert int to float
print(float_i, type(float_i))

# adding two floats when they are added together, the result is a float
float_1 = 3.50
float_2 = 4.50
sum_float = float_1 + float_2
print(sum_float, type(sum_float))

j = 5
k = 2
result = j / k
print(result, type(result))


#float numnbers can be positive, negative or zero
m = 0.0
print(m, type(m))


```

### Naming variables
- You can name your **variables** so that they are *self-explanatory* 
- Use lower_case

```python
first_name = "John"
last_name = "Doe"       
full_name = first_name + " " + last_name
print("Full name:", full_name)
```

### Boolean variables


```python
student_status = True
print(type(student_status))
# print("Is the student enrolled?", student_status)
student_status = False
print("Is the student enrolled?", student_status)

# Boolean Logic
A = True
B = False
print("A AND B:", A and B)
print("A OR B:", A or B)                
print("NOT A:", not A)
print("NOT B:", not B)
print("A XOR B:", A ^ B)
```

### String Methods

```python
str_one = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
print(str_one) 

# .lower() method
print(".lower() converts a string to lowercase:", str_one, "-->", str_one.lower())

# slicing a string
str_two = "Python Programming is great!"
print("Slicing a string:", str_two[0:6]) # Output: Python
print("Slicing a string:", str_two[:6]) # without defining the lower-bound: Output: Python
print("Slicing a string:", str_two[7:18]) # Output: Programming
print("Slicing a string:", str_two[19:21]) # Output: is great!
print("Slicing a string:", str_two[22:28]) # Output: is great!
print("Slicing a string:", str_two[22:]) # without defining the upper-bound:  Output: is great!

firs_and_last_name = "John Doe"
print("First name:", firs_and_last_name[:4]) # Output: John 

# removing extra whitespace from a string
name = "   John Doe   "
print(name.strip())

# String concatenation
first_name = "John" 
last_name = "Doe"
full_name = first_name + last_name        
print("Full name without space:", full_name) # Output: JohnDoe

full_name = first_name + " " +  last_name        
print("Full name with space:", full_name) # Output: John Doe
```

### F-String: Very powerfull way for formatting strings 

```python
age = 36
#This will produce an error:
txt = "My name is John, I am " + str(age)
print(txt)
```

```python
Name = "William"
age = 43

# Todo: create a statement like "Hello, I am John and I am 36 years old."
print("Hello, I am " + Name + " and I am " + str(age) +" years old.")  # Concatination
print("Hello, I am" , Name , "and I am" , str(age) , "years old.")     # Comma-separated values, which will automatically add spaces  


# # let's covnvert the above to an f-string, which is a more modern and efficient way to format strings in Python
print("Let'see the F-string version of the above statement:")
print(f"Hello, I am {Name} and I am {age} years old.")
```
