In this JN, we will explore variables and their different types

### Integer Variables and how to add them

```python
a = 10
b = 25
c = a + b
print("The sum of", a, "and", b, "is", c)   
```

### Naming variables
- You can name your **variables** so that they are *self-explanatory* 
- Use lower_case

```python
first_name = "John"
last_name = "Doe"       
full_name = first_name + " " + last_name
print("Full name:", full_name)
```

### Boolean variables


```python
student_status = True
print(type(student_status))
# print("Is the student enrolled?", student_status)
student_status = False
print("Is the student enrolled?", student_status)

# Boolean Logic
A = True
B = False
print("A AND B:", A and B)
print("A OR B:", A or B)                
print("NOT A:", not A)
print("NOT B:", not B)
print("A XOR B:", A ^ B)
```
