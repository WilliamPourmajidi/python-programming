### is and is not V.S  ==  and !=





```python
my_first_list = [1, 2, 3, 4, 5]
my_second_list = ["apple", "banana", "cherry"]
my_third_list = [1, 2, 3, 4, 5]
my_forth_list = my_first_list 

print(my_first_list is my_second_list) # False , they are different objects in memory
print(my_first_list == my_second_list) # False , they have different values
print("---" * 5)
print(my_first_list is my_third_list) # False
print(my_first_list == my_third_list) # True
print("---" * 5)
print(my_first_list is my_forth_list) # True
print(my_first_list == my_forth_list) # True
```

```python

# Let's check if we change my_first_list, will it affect my_forth_list?

my_first_list.append(6)
print(f"my_first_list = {my_first_list}") 
print(f"my_forth_list = {my_forth_list}")  # please note that my_forth_list is also changed because it is referencing the same object in memory as my_first_list
```

```python
a = 10
b = 10  

if (a == b):
    print("a is equal to b")        
else:
    print("a is not equal to b")


if (a != b):
    print("a is not equal to b")
else:
    print("a is equal to b")    


```

### String comparisons:  are inherently performed character by character based on the Unicode code points

```python
my_first_string = "Hello"
my_second_string = "Hello"
print(my_first_string == my_second_string) # True, because they have the same value
```

```python
### ASCII value of a character
char = 'h'
ascii_value = ord(char)
print(f"The ASCII value of '{char}' is {ascii_value}")
```

```python
my_long_string = "Hello World!"


# Let's go character by character in the string and print the ASCII value of each character
for char in my_long_string:
    ascii_value = ord(char)
    print(f"The ASCII value of '{char}' is {ascii_value}")
```
