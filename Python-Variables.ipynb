```python
# Christmas tree using * characters
height = 6

# Tree leaves
for i in range(height):
    stars = "*" * (2 * i + 1)
    print(stars.center(2 * height - 1))

# Tree trunk
trunk_width = 3
trunk_height = 3
for _ in range(trunk_height):
    print(("*" * trunk_width).center(2 * height - 1))
```

### F-String: Very powerfull way for formatting strings 

```python
age = 36
#This will produce an error:
txt = "My name is John, I am " + str(age)
print(txt)
```

```python
Name = "William"
age = 43

# Todo: create a statement like "Hello, I am John and I am 36 years old."
print("Hello, I am " + Name + " and I am " + str(age) +" years old.")  # Concatination
print("Hello, I am" , Name , "and I am" , str(age) , "years old.")     # Comma-separated values, which will automatically add spaces  


# # let's covnvert the above to an f-string, which is a more modern and efficient way to format strings in Python
print("Let'see the F-string version of the above statement:")
print(f"Hello, I am {Name} and I am {age} years old.")
```

### Assignment Operatins



```python
x = 0
x = 10 # this is an assignment example 
x = x + 5 # this is an augmented assignment example, it adds 5 to the current value of x and assigns the result back to x
x = x - 2 # this is another augmented assignment example, it subtracts 2 from the current value of x and assigns the result back to x
 
```

```python

# more common way to write the above arugements
x = 20
x += 5 # this is the augmented assignment operator for addition, it adds 5 to the current value of x and assigns the result back to x
x -= 2 # this is the augmented assignment operator for subtraction, it subtracts 2 from the current value of x and assigns the result back to x
print(x)
```

### Generating Random Numbers

```python
import random

# Random integer between 1 and 100 (inclusive)
rand_int = random.randint(1, 100)
print("Random integer:", rand_int)

# # Random float between 0 and 1
rand_float = random.random()
print("Random float:", rand_float)

# # Random float in a custom range
rand_uniform = random.uniform(10, 20)
print("Random float between 10 and 20:", rand_uniform)

# # Random choice from a list
colors = ["red", "green", "blue", "yellow"]
print("Random color:", random.choice(colors))
```

```python
my_stuents = ["Alice", "Bob", "Charlie", "David", "Eve", "Simrandeep" , "William", "Manpreet"]
volunteer = random.choice(my_stuents)   
print("Randomly selected volunteer:", volunteer)
```
