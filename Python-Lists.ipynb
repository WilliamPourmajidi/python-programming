- Slicing uses an **exclusive upper bound**: `my_list[a:b]` includes indices `a` up to `b-1`.

## Solution 2A

````
banner("Solution 2A")
a = [1, 2, 3]
b = a
b.append(99)
print("a =", a)
print("b =", b)
print("Observation: a changed because b is an alias of a.")
````

## Solution 2B

````
banner("Solution 2B")
a = [1, 2, 3]
c = a.copy()   # or a[:]
c.append(99)
print("a =", a)
print("c =", c)
print("Observation: a did not change because c is a copy.")
````
