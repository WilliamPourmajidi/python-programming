```python
# Passing multiple arguments to a function
def full_name(first_name, last_name):
    return f"Your first name is {first_name} and  your last name is {last_name} and your full name is {first_name} {last_name}"

my_record = full_name("William", "Smith")
print(my_record)
```

```python
# A boolean function
def is_even(num):
    if num % 2 == 0:
        return True 
    else:
        return False
    

is_even_result = is_even(222)
print(is_even_result)
```
